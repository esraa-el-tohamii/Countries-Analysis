```python
import sqlite3
import pandas as pd
```

```python
file='database.sqlite/database.sqlite'
conn=sqlite3.connect(file)
```

```python
query="""SELECT *
FROM sqlite_master
WHERE TYPE='table'"""
result=pd.read_sql(query,conn)
result
```

```python
query="""SELECT *
FROM Indicators
LIMIT 10
"""
result=pd.read_sql(query,conn)
result
```

```python
query="""SELECT COUNT(*)
FROM Indicators
WHERE IndicatorName=NULL
"""
result=pd.read_sql(query,conn)
result
```

```python
query="""SELECT COUNT(*)
FROM Indicators
WHERE IndicatorName is NOT NULL
"""
result=pd.read_sql(query,conn)
result
```

```python
query="""SELECT COUNT(*)-COUNT (CountryName) AS diff_country_name,
COUNT(*)-COUNT (CountryCode) AS diff_country_code,
COUNT(*)-COUNT (IndicatorName) AS diff_indicator_name,
COUNT(*)-COUNT (IndicatorCode) AS diff_indicator_code,
COUNT(*)-COUNT (Year) AS diff_year,
COUNT(*)-COUNT (Value) AS diff_value
FROM Indicators
where IndicatorName = "CO2 emissions (kt)"
"""
result=pd.read_sql(query,conn)
result
```

```python
query = """Select * from Indicators
where IndicatorName like '%ratio%'"""
result = pd.read_sql(query,conn)
result
```

```python
query = """select distinct(IndicatorName)
from Indicators
where IndicatorName like 'GDP%'"""
result = pd.read_sql(query,conn)
result
```

```python
query = """select * from Indicators
where IndicatorName = 'GDP (constant LCU)' and (CountryName = 'West Bank and Gaza' or Year = 1960)"""
result = pd.read_sql(query,conn)
result
```

```python
query = """select * 
from Indicators
where IndicatorName = 'GDP (constant LCU)' and (CountryName = 'West Bank and Gaza' or Year between 1960 and 2000 )"""
result = pd.read_sql(query,conn)
result
```

```python
query="""select CountryName , avg(Value)
from Indicators 
where IndicatorName = 'GDP (constant LCU)' and CountryName in ('Algeria','Argentina','Australia')
GROUP BY CountryName"""
result = pd.read_sql(query,conn)
result
```

```python
query="""select CountryName , Value,CASE
WHEN Value<10000 THEN 'Low'
WHEN Value>80000 THEN 'High'
ELSE 'Medium'
END category
from Indicators 
where IndicatorName = 'GDP (constant LCU)' """
result = pd.read_sql(query,conn)
result
```

```python
query="""select *,AVG(Value) OVER(partition by CountryCode)
from Indicators
LIMIT 1000
"""
result = pd.read_sql(query,conn)
result
```
